```python
import pandas as pd
import time

# Load Excel file
df = pd.read_csv("C:/Users/baner/Downloads/sales_data.csv")
df.head()

```

```python
# Start timer for the automated process
start_time_automated = time.time()

# Clean the data
df_cleaned = df.dropna()
df_cleaned = df_cleaned.drop_duplicates()

df_cleaned

```

```python
# Format columns
df_cleaned['Sales_Date'] = pd.to_datetime(df_cleaned['Sales_Date'], errors='coerce')
```

```python
# Extract month and year from Date column
df_cleaned['Month'] = df_cleaned['Sales_Date'].dt.to_period('M')
```

```python
# Create Pivot Tables
sales_by_region = pd.pivot_table(df_cleaned, values='Sales_Amount', index='Region', aggfunc='sum')
monthly_totals = pd.pivot_table(df_cleaned, values='Sales_Amount', index='Month', aggfunc='sum')

```

```python
# Save cleaned data and pivot tables to a new Excel file
with pd.ExcelWriter('cleaned_sales_data.xlsx', engine='openpyxl') as writer:
    df_cleaned.to_excel(writer, sheet_name='Cleaned Data', index=False)
    sales_by_region.to_excel(writer, sheet_name='Sales by Region')
    monthly_totals.to_excel(writer, sheet_name='Monthly Totals')

```

```python
# End timer for the automated process
end_time_automated = time.time()

```

```python
# Time for the automated process
automated_time = end_time_automated - start_time_automated
print(f"Time for automated process: {automated_time:.2f} seconds")


```
